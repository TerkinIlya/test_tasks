```python
def get_all_articles(url, cookies, headers):
  all_articles = []
  for page in range(1, 37):
    response = requests.get(url, cookies=cookies,
                          headers=headers, params={
    'category': 'gold-bracelets-with-stones',
    'sort': 'popular',
    'page': f'{page}'})
    for res in response.json()['results']:
      all_articles.append(res.get('article', None))
  return all_articles
```

```python
all_articles = get_all_articles(url=url, cookies=cookies, headers=headers)
```

```python
def get_all_products(articles, cookies, headers):
  products = {'name': [],
        'price': [],
        'online_price': [],
        #'weight': [],
        'share_url': []}
  for art in all_articles:
    response = requests.get(f"https://www.585zolotoy.ru/api/products/{art}/",
                            cookies=cookies, headers=headers).json()
    products['name'].append(response.get('name', None))
    products['price'].append(float(response.get('price')))
    products['online_price'].append(response.get('online_price'))
    #products['weight'].append(response.get('properties',
    #                                       None)[1]['values'][0][2]['value'][0])
    products['share_url'].append(response.get('share_url', None))
  return products
```

```python
products = get_all_products(all_articles, cookies, headers)
```

```python
bracelets = pd.DataFrame(products)
bracelets
```
